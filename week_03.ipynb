### 논리연산자
- 비교연산자가 여러번 사용될 때 사용하는 것
- and, or, not
    - A and B / A & B : A도 참이고, B도 참일 때 True 
    - A or B / A | B : A 혹은 B, 둘 중 하나만 참이며 True
    - not A : A가 False -> True, True -> False
- 비교연산자의 괄호는 가독성을 위해 (선택 가능)
- 파이썬에서만 가능한 표현 : 0 < x < 10

### True, False
- "만약 a가 참이라면 ..."
- a = True

````python
a = True
if a == True:
    print(1) 
#틀림! 아래와 같이 써야함.

if a:
    print(1)
````

### in

- membership operator
- 어떤 요소가 그 안에 있는지 확인하고 싶다.
- 비슷한 조건이 여러번 반복될 때 사용 가능


````python
letter = 'k'

# letter에 모음이 있는지 찾고싶음.

# if letter == 'a' or letter === 'e' ....in을 사용하지 않으면 코드가 복잡해짐.
if letter in ['a','e','i','o','u']:
    print("True")
else:
    print("No")
````

### 바다코끼리 연산자
- 할당과 테스트를 동시에 하는 기능
- walrus operator
- :=

````python
tweet_limit = 200
tweet_string = 'blach'*50 #공백 없이 스트링 4*50개
tweet_string
````

````python
diff = tweet_limit - len(tweet_string)
if diff>0:
    print("O")
else:
    print("X")
````

````python
if diff := tweet_limit - len(tweet_string) > 0:
        print("O")
````

### Quiz
- 윤년을 구하는 코드를 짜보자
    - 연도가 4로 나눠떨어지면 윤년
    - 100으로 나눠떨어지면 안됨.
    - 400으로 나눠떨어지면 된다.
    
- year를 입력 받는다.
- '{}년은 윤년입니다' / '{}년은 윤년이 아닙니다' 

````python
#원래 코드
year = int(input())

if year%4==0:
    if year%100==0:
        if year%400==0:
            print(f"{year}년은 윤년입니다")
        else:
            print(f"{year}년은 윤년이 아닙니다")
    else:
        print(f"{year}년은 윤년입니다")
````

````python
#더 나은 코드
year = int(input())

if year%4==0 and year%100!=0 or year%400:
    print(f"{year}년은 윤년입니다")
else:
    print(f"{year}년은 윤년입니다")
````

0 1 2 ...<br>

        -1

#### NLTK
- natural language toolkit
- 교육용으로 개발된 자연어처리 패키지

### NLTK란?
- 자연어 처리를 위한 파이썬의 패키지
        - 자연어란 ? 우리가 일상생활에서 사용하는 언어.
        - 자연어 처리란? 컴퓨터가 자연어를 처리할 수 있도록 하는 것.
        - 자연어 처리가 되면, 이 정보를 바탕으로 음성 인식, 내용 요약, 감성 분석, 텍스트 분류 등을 할 수 있음.

    - 1. 앞에서 단어 10개 추출하기
    - 조건문 연습 : 'z' 가 있으면 대문자로 바꾸기, 'z'가 없는데 4자 이상이면 끝 글자 대문자로 바꾸기

- 주민등록번호, 휴대폰번호, 이메일 입력 받는다. 
    - 주민번호, 휴대폰은 숫자만 입력(-없다)
    - 이메일은 아이디만 입력(@ 뒷부분 없음)
- 출력 : 
    - 당신은 {}년 {}월 {}일 출생의 {남성/여성}입니다.
    - 휴대폰 번호 : {} 하이픈으로 연결하기
    - 이메일주소 {아이디}@gmail.com

````python
thu = book.text9
len(set(thu.tokens))/len(thu.tokens)
````

````python
pyt = book.text6
pyt_set = sorted(set(pyt.tokens),reverse=True)
print(pyt_set[:10])

result_pyt_set = []

for word in pyt_set:
    if 'z' in word:
        word = word.replace('z','Z')
    else:
        if len(word)>=4:
            word = word[:-1]+word[-1].upper()
    result_pyt_set.append(word)
            
for word in result_pyt_set:
    print(word)
        
````
